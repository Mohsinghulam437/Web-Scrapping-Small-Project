```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page_count = 1
while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No More Valid Pages.")
        break
    with open(f"scrap/page{page_count}.html", "w", encoding = "utf-8") as f:
        f.write(res.text)
        print(f"Download Data from {page_count}")

    page_count += 1
```

```python
page_count = 1
total_quotes = []
while page_count <=10:
    with open(f"scrap/page{page_count}.html", "r", encoding = "utf-8") as f:
        content = f.read()
    soup = BeautifulSoup(content, "lxml")
    quote = soup.select("div.quote")

    for q in quote:
        tags_list = []
        tags = q.select(".tags .tag")
        for tag in tags:
            tags_list.append(tag.get_text())
        if "life" in tags_list:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            total_quotes.append([text, author])
            print(f"Data Saved Successfully in Page No.{page_count}")
    page_count += 1
    
        
```

```python
quotes = pd.DataFrame(total_quotes, columns = ("Quote", "Author"))
quotes.to_csv("life_quotes.csv", index = False)
```
